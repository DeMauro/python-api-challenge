### Humidity Heatmap

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
Hotel_Name = [] # empty list to append and add to the dataframe
target_radius = 5000
target_search = 'Hotel'
target_type = "lodging"
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# some of my places did not have hotels within 5,000 meters for example nalut in Libya
# https://www.google.com/search?q=nalut%2C+LY&oq=nalut%2C+LY&aqs=chrome..69i57.2640j0j1&sourceid=chrome&ie=UTF-8
# The try/except loop changes 5,000 meters to rankBy:distance.  But even then some don't have hotels nearby, 
# so I add NaN to the hotel name and remove them from the list

for index, row in hotel_df.iterrows():
    # target coord of each city in the list
    lat = row['Lat']
    lng = row['Lng']
    target_coordinates = f"{lat},{lng}"  
    try:
        params = {"location": target_coordinates, "language":"en","region":"us","keyword": target_search,
                  "radius": target_radius,"type": target_type,"key": g_key}
        response = requests.get(base_url, params=params).json()
        Hotel_Name.append(response['results'][0]["name"])
    except:
        params = {"location": target_coordinates, "language":"en","region":"us","keyword": target_search,
                  "rankBy":"distance","type": target_type,"key": g_key}
        response = requests.get(base_url, params=params).json()
        if response['status']=='INVALID_REQUEST':
            Hotel_Name.append('NaN')
        else:
            Hotel_Name.append(response['results'][0]["name"])           
hotel_df["Hotel Name"] = Hotel_Name
```

```python
# dropping places that didn't have a hotel from abve
hotel_df = hotel_df[hotel_df['Hotel Name'] != 'NaN']
hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map

#from the class work files and https://jupyter-gmaps.readthedocs.io/en/latest/api.html
markers = gmaps.marker_layer(locations,info_box_content=hotel_info)
fig.add_layer(markers)
# Display figure
fig
```
